```python
import itertools
from collections import defaultdict
import sys
from queue import PriorityQueue
# import Queue as queue
import copy
import sys

#Function for reading the contraints file
def read_file(input_file):
    file = open(input_file, "r")
    data_dict = defaultdict(list)

    for line in file:
        if line != "\n":
            elements =  line.split(" ")
            name = elements[0].strip()
            size = elements[1].strip()
            like = elements[2].strip()
#             dislike = elements[3].strip()

            data_dict[name].append(size)
            data_dict[name].append(like)
#             data_dict[name].append(dislike)

    return data_dict
```

```python
def addNewGroup(person_list):

    groups = []

    if len(person_list)>3:

        for comb in itertools.combinations(person_list,3):
            groups.append(comb)

    elif len(person_list) > 2:

        for comb in itertools.combinations(person_list, 2):
            groups.append(comb)

    elif len(person_list)==2 or len(person_list)==3 or len(person_list) == 1:
        groups.append(tuple(person_list))

    return groups
```

```python
def successors_ucs(s):
    listOfSuccessors = []
    person_list = addNewGroup(s)
    possibleNewGroups = addNewGroup(person_list)
    for newgroups in possibleNewGroups:
        tempGroup = copy.deepcopy(s[1])
        if newgroups not in s[1]:
            tempGroup.append(newgroups)
            listOfSuccessors.append(tempGroup)

#     print("successors_ucs ====>  ")
#     print(listOfSuccessors)
    return listOfSuccessors
```
